```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable

s1 = "A child in a blue and white team uniform chasing a soccer ball .".split()
s2 = "A young boy in a blue soccer uniform chasing a ball .".split()
s1 = s1 + s1
s2 = s2 + s2
heatmap = np.arange(len(s1) * len(s2)).reshape(len(s1), len(s2)) / (len(s1) * len(s2))

def plot_heatmap(s1, s2, data, fpath):
    l1, l2 = len(s1), len(s2)

    fig, ax = plt.subplots(figsize=(int(0.5*len(s2)), int(0.3*len(s1))), facecolor='white')
    
    plt.gca().invert_yaxis()
    ax.set_yticks([y+0.5 for y in range(l1)])
    ax.set_yticklabels(s1)
    ax.set_xticks([x+0.5 for x in range(l2)])
    ax.set_xticklabels(s2, rotation=90)
    ax.xaxis.set_ticks_position('top')
    im = plt.pcolormesh(data, edgecolors='k', linewidths=1, cmap=plt.get_cmap("Blues"))
    for y in range(data.shape[0]):
        for x in range(data.shape[1]):
            plt.text(x + 0.5, y + 0.5, '%.2f' % data[y, x],
                    horizontalalignment='center',
                    verticalalignment='center',
                    c='w' if data[y, x] > 0.5 else 'k',
                    )

    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size="5%", pad=0.05)
    plt.colorbar(im, cax=cax)
    plt.tight_layout()
    plt.savefig(fpath, dpi=400)

plot_heatmap(s1, s2, heatmap, "test.png")
```
