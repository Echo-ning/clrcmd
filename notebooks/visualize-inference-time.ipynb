```python
import pandas as pd
df = pd.read_csv("/nas/home/sh0416/experimental-results/cl-rwmd/experiment - exp3. inferencetime.csv")
df.head()
```

```python
simcse_avg_result = df[df["method"] == "simcse-avg"]
simcse_cls_result = df[df["method"] == "simcse-cls"]
clrcmd_result = df[df["method"] == "rwmdcse"]
```

```python
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 16})
fig, ax = plt.subplots(facecolor='white')
clrcmd = ax.plot(clrcmd_result["seq len"], clrcmd_result["elapsed time"], 'ro-', label="CLRCMD")
simcse_avg = ax.plot(simcse_avg_result["seq len"], simcse_avg_result["elapsed time"], 'bo-', label="SimCSE-avg")
simcse_avg = ax.plot(simcse_cls_result["seq len"], simcse_cls_result["elapsed time"], 'go-', label="SimCSE-cls")
ax.set_xlabel("sequence length (# of tokens)")
ax.set_ylabel("elapsed time (ms)")
ax.legend()
plt.tight_layout()
plt.savefig("inference_time.png", dpi=200)
```
